```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
####Attrition by job role(Department)
plt.figure(figsize=(10,5))
sns.countplot(x="Department", hue="Attrition",data = data)
```

```python
####distance from home by attrition
plt.figure(figsize=(10,5))
sns.histplot(x='DistanceFromHome' ,hue='Attrition' , data=data, element='poly',edgecolor='white')
plt.grid()
```

# Q1.Show a breakdown of distance from home by job role(Department) and attrition.


```python
data.groupby(['Department','Attrition']).DistanceFromHome.mean().unstack()
```

```python
sns.boxplot(x="Department", y="DistanceFromHome", hue="Attrition", data=data)
```

# Q2.Compare average monthly income by education and attrition.

```python
data.groupby(['Education','Attrition']).DailyRate.mean().unstack()

```

```python
#Visualization
sns.boxplot(x="Education", y="DailyRate", hue="Attrition", data=data)
```
